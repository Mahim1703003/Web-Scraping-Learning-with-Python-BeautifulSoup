```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html = urlopen("https://keithgalli.github.io/web-scraping/webpage.html")
bsObj = BeautifulSoup(html.read())

table_tr_list = bsObj.find("table",{"class":"hockey-stats"}).findAll("tr")

for i in range(0,len(table_tr_list)):
    
    if i==0:
        table_tr_th_list = table_tr_list[i].findAll("th")
        for th_tag in table_tr_th_list:
            print(th_tag.get_text().strip(),end="\t")
    else:
        table_tr_td_list = table_tr_list[i].findAll("td")
        for td_tag in table_tr_td_list:
            print(td_tag.get_text().strip(),end="\t")
    
    print()
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd

html = urlopen("https://keithgalli.github.io/web-scraping/webpage.html")
bsObj = BeautifulSoup(html.read())

column_tags = bsObj.find("table",{"class":"hockey-stats"}).find("thead").find("tr").findAll("th")
column_names = list()

print(column_tags)

for column_tag in column_tags:
    column_names.append(column_tag.get_text())

print(column_names)

print()

row_tags = bsObj.find("table",{"class":"hockey-stats"}).find("tbody").findAll("tr")
rows = list()

for row_tag in row_tags:
    single_row = row_tag.findAll("td")
    row = list()
    
    for i in single_row:
        row.append(i.get_text().strip())
    
    rows.append(row)

print(rows)
print()

df = pd.DataFrame(rows,columns = column_names)
print(df)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://keithgalli.github.io/web-scraping/"

html = urlopen(url+"webpage.html")
bsObj = BeautifulSoup(html.read())

a_tags1 = bsObj.find("div",{"class":"block","align":"left"}).findAll("a")
a_tags2 = bsObj.find("div",{"class":"block","align":"center"}).findAll("a")

links = list()

for a_tag in a_tags1:
    links.append(a_tag.attrs['href'])

for a_tag in a_tags2:
    links.append(a_tag.attrs['href'])
    
secret_words = list() 

for link in links:
    html2 = urlopen(url+link)
    bsObj2 = BeautifulSoup(html2.read())
    
    p_tags = bsObj2.findAll("p",{"id":"secret-word"})
    
    for p_tag in p_tags:
        secret_words.append(p_tag.get_text())
        
for word in secret_words:
    print(word,end=" ")
    
print()
```
