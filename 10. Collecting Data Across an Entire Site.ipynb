```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

visitedLinks = list()

def getLinks(pageUrl):
    global visitedLinks
    
    html = urlopen("http://en.wikipedia.org"+pageUrl)
    bsObj = BeautifulSoup(html.read())
    
    try:
        print("h1 :",bsObj.find("h1").get_text())
        print("paragraph tag : ",bsObj.find("div",{"id":"mw-content-text"}).find("p"))
        print("edit link : ",bsObj.find("li",{"id":"ca-edit"}).find("a").attrs['href'])
    except:
        print("This page is missing something! No worries though")
        
    print()
    print("=================================================================================")
    print()
    
    links = bsObj.findAll("a",{"href":re.compile("^(\/wiki\/)")})
    
    for link in links:
        if link.attrs['href'] not in visitedLinks:
            newPage = link.attrs['href']
            print("New Page : ",newPage)
            visitedLinks.append(newPage)
            getLinks(newPage)
            
getLinks("")
```
